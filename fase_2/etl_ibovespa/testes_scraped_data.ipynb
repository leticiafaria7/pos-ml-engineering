## Imports

```python
import pandas as pd
import numpy as np
```

```python
from os import walk
from typing import List

def listar_arquivos_pasta(folder_path: str) -> List[str]:

    files = []
    for (dirpath, dirnames, filenames) in walk(folder_path):
        files.extend(filenames)
        break

    return files
```

## Ler dados

```python
arquivos = listar_arquivos_pasta('scraped/')
```

```python
pd.read_csv(f"scraped/{arquivos[0]}")
```
