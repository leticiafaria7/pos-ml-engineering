## Operações com listas

```python
frutas = ['maçã', 'laranja']
frutas.insert(1, 'banana')
print(frutas)

frutas.remove('banana') # remove a primeira ocorrência de banana
print(frutas)
```

```python
itens = ['a', 'b', 'c', 'd']
removido = itens.pop(2) # pop recebe a POSIÇÃO, não o item em si
print(removido)
print(itens)
```

```python
itens.reverse()
print(itens)
```

## Conjuntos

```python
numeros = {1, 2, 3, 3, 2, 1}
print(f"Set de números: {numeros}")
```

```python
numeros.add(4)
numeros
```

```python
numeros.discard(2)
numeros
```

```python
conjunto1 = {1, 2, 3}
conjunto2 = {3, 4, 5}

uniao = conjunto1.union(conjunto2)
print(f'União: {uniao}')
```

```python
intersecao = conjunto1.intersection(conjunto2)
print(f"Interseção: {intersecao}")
```

```python
diferenca = conjunto1.difference(conjunto2)
print(f"Diferença: {diferenca}")
```
