Os Objetivos são que cada tarefa possua: título, descrição e data de conclusão. 
Além disso, você deve criar métodos para marcar as tarefas como concluídas, adicionar tarefas, remover tarefas e listar as tarefas do gerenciador.

```python
import datetime as dt
import pandas as pd
```

```python
data_hora_conclusao = pd.to_datetime(dt.datetime.now())
dia_semana = data_hora_conclusao.day_name()

print(f"Data de conclusão: {data_hora_conclusao.strftime('%d/%m/%Y %H:%M')} ({dia_semana})")
```

```python
class Tarefa:

    _lista_tarefas = {}

    def __init__(self, titulo:str, descricao:str):
        self.titulo = titulo
        self.descricao = descricao
        self.data_inicio = pd.to_datetime(dt.datetime.now()).strftime('%d/%m/%Y %H:%M')
        self.data_conclusao = '-'
        self.concluida = 'Não'
        
        index = len(Tarefa._lista_tarefas) + 1
        self.index = index
        Tarefa._lista_tarefas[index] = {}

        Tarefa._lista_tarefas[index]['Tarefa'] = self.titulo
        Tarefa._lista_tarefas[index]['Descrição'] = self.descricao
        Tarefa._lista_tarefas[index]['Data de início'] = self.data_inicio
        Tarefa._lista_tarefas[index]['Concluída'] = self.concluida
        Tarefa._lista_tarefas[index]['Data de conclusão'] = self.data_conclusao
    
    def __str__(self):

        return f"""
#:                 {self.index}
Tarefa:            {self.titulo}
Descrição:         {self.descricao}
Data de início:    {self.data_inicio}
Concluída?         {self.concluida}
Data de conclusão: {self.data_conclusao}
"""

    def concluir_tarefa(self):
        self.data_conclusao = pd.to_datetime(dt.datetime.now()).strftime('%d/%m/%Y %H:%M')
        self.concluida = 'Sim'
        Tarefa._lista_tarefas[self.index]['Concluída'] = self.concluida
        Tarefa._lista_tarefas[self.index]['Data de conclusão'] = self.data_conclusao

    def remover_tarefa(self):
        check_remocao = input(f"Tem certeza de que quer remover a tarefa {self.index}? (s/n)")

        if check_remocao == 's':
            Tarefa._lista_tarefas.pop(self.index)
        else:
            pass

    def lista_tarefas(ver_tabela = True):
        if ver_tabela:
            print(f"# | {'Título'.ljust(10)} | {'Descrição'.ljust(70)} | {'Data de início'.ljust(20)} | Concluída? | Data de conclusão")
            print('-' * 150)
            for k, v in Tarefa._lista_tarefas.items():
                print(f"{k} | {v['Tarefa'].ljust(10)} | {v['Descrição'].ljust(70)} | {v['Data de início'].ljust(20)} | {v['Concluída'].ljust(10)} | {v['Data de conclusão']}")

        # return Tarefa._lista_tarefas
```

```python
tarefa1 = Tarefa(titulo = 'Tarefa 1', descricao = 'Criar classe tarefa com função de concluir tarefa')
print(tarefa1.__str__())
```

```python
Tarefa.lista_tarefas()
```

```python
tarefa1.concluir_tarefa()
print(tarefa1.__str__())
```
