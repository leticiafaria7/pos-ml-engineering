## Importações

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# configurar opções do chrome
chrome_options = Options()
chrome_options.add_argument('--headless') # sem interface gráfica
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.7499.40 Safari/537.36"
)
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
chrome_options.add_experimental_option('useAutomationExtension', False)
```

```python
# inicializar o WebDriver com o WebDriver Manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service = service, options = chrome_options)
```

## Automatizar uma pesquisa no Google

```python
# abrir o google
driver.get('https://www.google.com')

# encontrar a barra de pesquisa
search_box = driver.find_element(By.NAME, 'q')

# digitar a consulta e enviar
search_query = 'Pos Tech FIAP'
search_box.send_keys(search_query)
search_box.submit()

# aguardar os resultados carregarem
wait = WebDriverWait(driver, 15)

# --- tenta clicar no botão de aceitar cookies ---
try:
    accept_btn = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, "//button[contains(., 'Aceitar')] | //div[contains(text(),'Aceitar tudo')]"))
    )
    accept_btn.click()
except:
    pass

# capturar os títulos de resultados
try:
    titles = driver.find_elements(By.CSS_SELECTOR, 'div#search div.MjjYud h3')
except:
    # Se o seletor acima falhar, tente o genérico
    titles = driver.find_elements(By.TAG_NAME, 'h3')

print(f"Resultados para a pesquisa: '{search_query}'\n")
if titles:
    for idx, title in enumerate(titles, start = 1):
        # O resultado do título da pesquisa em si (o texto da pesquisa) é geralmente
        # o primeiro h3, pule ele se necessário ou filtre por texto não vazio.
        if title.text:
            print(f"{idx}. {title.text}")

    print(f"\nTotal de resultados (h3s) encontrados: {len(titles)}")
else:
    print("Nenhum resultado encontrado com o seletor atual. O seletor pode precisar de atualização.")

# fechar o navegador
driver.quit()
```
